```python
import os
import pandas as pd
from nnunet.evaluation.evaluator import evaluate_folder
# some constant
TASK_ID = 500
INPUT_DIR = "/local/nnguye02/dataset/nnUNet_raw_data/Task500_Epaule/imagesTs"
LABEL_TEST_DIR = "/local/nnguye02/dataset/nnUNet_raw_data/Task500_Epaule/labelsTs"
OUTPUT_DIR = "/local/nnguye02/dataset/nnUNet_inference_test"
```

# For models 2d, 3d_fullres, TransUNet2D, TransUNet3D

To run inference on nifti images with nnUNet, we have to specify:
- input folder: must contain nifti files for each patient
- output folder: predictions will be saved here
- task name or task ID

The prediction can take time based on the number of images and the chosen model.

```python
model = "TransUNet2D" # MODIFY IF USE OTHER MODEL
assert model in ["2d", "3d_fullres", "3d_cascade_fullres", "TransUNet2D", "TransUNet3D"]

pred_dir = os.path.join(OUTPUT_DIR, model)
```

```python
!nnUNet_predict -i {INPUT_DIR} -o {pred_dir} -t {TASK_ID} -m {model}
```

# Evaluate predictions
After having predictions, we can run evaluation between the predictions and groundtruths.

Hausdorff Distance and Average Symmetric Surface Distance are in milimeters.

```python
labels = [1, 2, 3, 4, 5]
summary = evaluate_folder(LABEL_TEST_DIR, pred_dir, labels, overwrite_summary=False)
mean = summary["results"]["mean"]
pd.DataFrame(mean)
```
